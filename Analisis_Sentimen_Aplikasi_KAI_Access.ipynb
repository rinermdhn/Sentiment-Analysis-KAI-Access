### Case-Folding
Tahapan *case-folding* mengubah seluruh huruf pada ulasan menjadi huruf kecil. Tahapan ini diperlukan sebelum pengolahan lebih lanjut karena Python merupakan bahasa pemrograman dengan sifat *case-sensitive*.

```python
# Case-folding Step

def lowercase(review_text):
  low = review_text.lower()
  return low

df['clean_review'] = df['content'].apply(lambda low:lowercase(str(low)))
```

### Data Cleaning
Tahapan *data cleaning* melibatkan proses penghapusan emoji, *hashtag*, angka, tanda baca, repetisi kata, dan spasi berlebih.

```python
# Remove Emoji Step

def remove_emoji(review_text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           u"\U0001f926-\U0001f937"
                           u"\U00010000-\U0010ffff"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', review_text)

df['clean_review'] = df['clean_review'].apply(lambda emoji: remove_emoji(emoji))
```

```python
# Remove Hashtags

def remove_hashtag(review_text, default_replace=""):
  hashtag = re.sub(r'#\w+', default_replace, review_text)
  return hashtag

df['clean_review'] = df['clean_review'].apply(lambda hashtag: remove_hashtag(hashtag))
```

```python
# Remove Numbers

def remove_number(review_text, default_replace=" "):
  num = re.sub(r'\d+', default_replace, review_text)
  return num

df['clean_review'] = df['clean_review'].apply(lambda num: remove_number(num))
```

```python
# Remove Punctuation

def remove_punctuation(review_text, default_text=" "):
  list_punct = string.punctuation
  delete_punct = str.maketrans(list_punct,' '*len(list_punct))
  new_review = ' '.join(review_text.translate(delete_punct).split())

  return new_review

df['clean_review'] = df['clean_review'].apply(lambda punct: remove_punctuation(punct))
```

```python
# Remove Word Repetition
 ## Pembatasan jumlah huruf menjadi dua (kembali ke bentuk awal)
 ## Menghindari kata ganda dengan arti sama, tetapi beda penulisan

def word_repetition(review_text):
  review = re.sub(r'(.)\1+', r'\1\1', review_text)
  return review

df['clean_review'] = df['clean_review'].apply(lambda word: word_repetition(word))
```

```python
# Pembatasan kata berulang menjadi satu kali

def repetition(review_text):
  repeat = re.sub(r'\b(\w+)(?:\W\1\b)+', r'\1',review_text, flags=re.IGNORECASE)
  return repeat

df['clean_review'] = df['clean_review'].apply(lambda word: repetition(word))
```

```python
# Remove Extra Whitespaces

def remove_extra_whitespaces(review_text):
  review = re.sub(r'\s+',' ', review_text)
  return review

df['clean_review'] = df['clean_review'].apply(lambda extra_spaces: remove_extra_whitespaces(extra_spaces))
```

### Stemming
*Stemming* mengubah kata berimbuhan menjadi bentuk dasar menggunakan *library* Sastrawi. *Stemming* dilakukan sebelum proses *indexing* untuk menyederhanakan variasi kata.

```python
# Stemming Step

def back_to_root(review_text):
  stop_token = ' '.join(review_text)
  stem = StemmerFactory()
  create_stem = stem.create_stemmer()
  result_stem = create_stem.stem(stop_token)
  return result_stem

df['stem_review'] = df['stop_review'].apply(lambda stem:back_to_root(stem))
```

## Sentiment Analysis

Setelah dilakukan **preprocessing data**, data ulasan siap untuk diolah lebih lanjut. Tahapan berikutnya adalah membangun model untuk analisis sentimen. Data dibagi menjadi dua bagian, yaitu **data latih** dan **data uji** dengan proporsi pembagiannya adalah **80%** untuk data latih dan **20%** untuk data uji. Setelah itu, data latih dan data uji digunakan untuk mengevaluasi performa model klasifikasi.

```python
# Splitting Data

X_train, X_test, y_train, y_test = train_test_split(X_tfidf, df_new['sentimen'], test_size=0.2, random_state=42)
X_train.shape, X_test.shape
```

```python
# Model yang Digunakan

models = {
    "Random Forest": RandomForestClassifier(random_state=42, n_jobs=5),
    "Naive Bayes": MultinomialNB()
}
```

### Random Forest

```python
results = []

# 5-fold Cross-validation setting
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for model_name, model in tqdm(models.items(), desc="Training Models", unit="model"):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Cross-validation scores
    f1_cv = cross_val_score(model, X_train, y_train, cv=cv, scoring='f1_macro', n_jobs=-1).mean()
    precision_cv = cross_val_score(model, X_train, y_train, cv=cv, scoring='precision_macro', n_jobs=-1).mean()
    sensitivity_cv = cross_val_score(model, X_train, y_train, cv=cv, scoring='recall_macro', n_jobs=-1).mean()
    accuracy_cv = cross_val_score(model, X_train, y_train, cv=cv, scoring='accuracy', n_jobs=-1).mean()

    # Test set scores
    f1 = f1_score(y_test, y_pred, average='macro')
    precision = precision_score(y_test, y_pred, average='macro')
    sensitivity = recall_score(y_test, y_pred, average='macro')
    accuracy = accuracy_score(y_test, y_pred)

    results.append({
        "Model": model_name,
        "F1 Score Average (CV)": f1_cv,
        "F1 Score (Test Set)": f1,
        "Precision Average (CV)": precision_cv,
        "Precision (Test Set)": precision,
        "Sensitivity Average (CV)": sensitivity_cv,
        "Sensitivity (Test Set)": sensitivity,
        "Accuracy Average (CV)": accuracy_cv,
        "Accuracy (Test Set)": accuracy
    })

# Result
model_df = pd.DataFrame(results)
model_df
```

Berdasarkan *output* di atas, terlihat bahwa tidak terdapat perbedaan jauh dari performa kedua model, baik dengan menerapkan *cross-validation* (CV) maupun tanpa menerapkan *cross-validation* (Test Set). Akan tetapi, **Random Forest** memiliki nilai yang **lebih tinggi** dibandingkan **Naive Bayes** di seluruh metrik evaluasi, kecuali pada Precision.

```python
# Parameter for Random Forest
param_grid = {
    'n_estimators': [100, 300, 500],
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['sqrt', 'log2']
}

# Model Initiation
rf = RandomForestClassifier(random_state=42, n_jobs=-1)

# Grid Search
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    scoring='accuracy',
    cv=5,
    n_jobs=-1,
    verbose=1
)

grid_search.fit(X_train, y_train)

print("Best Accuracy:", grid_search.best_score_)
print("Best Parameters:", grid_search.best_params_)
```

Proses *hyperparameter tuning* menggunakan **Grid Search**, yaitu teknik optimasi yang digunakan untuk menemukan kombinasi hiperparameter terbaik. Grid Search memanfaatkan hasil *cross-validation* dan memberikan kemudahan dalam menguji setiap parameter model tanpa harus melakukan validasi manual satu per satu [12]. Nilai k yang digunakan adalah k = 5.

Saat **Grid Search** dijalankan untuk **Random Forest**, terdapat 144 kombinasi sehingga menghasilkan 720 model latih. Didapatkan **Best Parameters** sebagai berikut:

* `criterion`: gini
* `max_depth`: None
* `max_features`: log2
* `min_samples_leaf`: 1
* `min_samples_split`: 5
* `n_estimators`: 300

Model dengan parameter terbaik ini digunakan untuk memprediksi **sentimen** atau label pada **data uji**.


```python
# Best Random Forest Model from Grid Search
best_rf_model = grid_search.best_estimator_

# StratifiedKFold Initiation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Prediction with Cross-Validation
y_pred_cv = cross_val_predict(best_rf_model, X_train, y_train, cv=cv, n_jobs=-1)

# Classification Report and Confusion Matrix for Random Forest Model (CV)
print("=== Cross-Validation Classification Report (CV) ===")
print(classification_report(y_train, y_pred_cv))
print("=== Confusion Matrix (CV) ===")
print(confusion_matrix(y_train, y_pred_cv))

# Prediction for Test
y_pred_rf = best_rf_model.predict(X_test)

# Classification Report and Confusion Matrix for Random Forest Model
print("\n=== Cross-Validation Classification Report ===")
print(classification_report(y_test, y_pred_rf))
print("=== Confusion Matrix ===")
print(confusion_matrix(y_test, y_pred_rf))
```

Performa model dievaluasi berdasarkan **Confusion Matrix** dan metrik pada **Classification Report**, yaitu untuk **Precision**, **Recall**, dan **F1 Score** untuk masing-masing label. Selain itu, **Accuracy** secara keseluruhan juga dipertimbangkan untuk menilai kinerja model secara umum.

* **Recall** untuk kelas negatif adalah 0.92, hal ini menunjukkan bahwa dari semua data negatif yang seharusnya terdeteksi, 92% berhasil ditemukan. Sementara untuk positif, 85% berhasil ditemukan. Model lebih sensitif terhadap data negatif.
